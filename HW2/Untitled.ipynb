Задача 1. Группировка по анаграммам
Написать программу, которая принимает список слов и группирует их по анаграммам. Порядок слов в группах и порядок самих групп не имеет значения.

Подумайте, какую структуру данных здесь оптимально использовать

Анаграммы — это слова, состоящие из одних и тех же букв, но в разном порядке (например, "nos", "son", "ons").

Примеры:

words = ["eat", "tea", "tan", "ate", "nat", "bat"]
Результат: [['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat']]
words = ["nos", "son", "mac", "cam", "ons"]
Результат: [['nos', 'son', 'ons'], ['mac', 'cam']]
words = ["a"]
Результат: [['a']]

```python
def group_anagrams(words):
    groups = {}
    for word in words:
        key = ''.join(sorted(word))
        if key in groups:
            groups[key].append(word)
        else:
            groups[key] = [word]
    return list(groups.values())
input_words = input().split()
result = group_anagrams(input_words)
for group in result:
    print(group)
```

Задача 2. Шпионская Сеть
У вас есть доступ к логам доступа к секретным документам. Логи представлены в виде списка кортежей (имя_агента, номер_документа). Написать программу, которая вернет множество имен агентов, имеющих доступ ко всем документам из списка required_docs.

Примеры:

access_logs = [("Alice", "doc1"), ("Bob", "doc2"), ("Alice", "doc2"), ("Charlie", "doc1")]
required_docs = {"doc1", "doc2"}

Алиса имеет доступ к {"doc1", "doc2"}. Боб — к {"doc2"}. Чарли — к {"doc1"}.
Только Алиса имеет доступ ко всем требуемым документам.
Результат: {"Alice"}

******************************

access_logs = [("Agent_X", "plan_A"), ("Agent_Y", "plan_B"), ("Agent_X", "plan_B")]
required_docs = {"plan_A", "plan_B", "plan_C"}

Ни один агент не имеет доступа ко всем трем документам.
Результат: set() (пустое множество)

******************************

access_logs = [("Anna", "file1"), ("Boris", "file2"), ("Anna", "file3"), ("Cyril", "file1"), ("Anna", "file2")]
required_docs = {"file1", "file2"}

Анна имеет доступ к {"file1", "file2", "file3"}. Борис — к {"file2"}. Кирилл — к {"file1"}.
Доступ Анны ({"file1", "file2", "file3"}) включает в себя все требуемые документы ({"file1", "file2"}).
Результат: {"Anna"}

```python
def find_agents(access_logs, required_docs):

    agent_docs = {}

    for agent, doc in access_logs:
        if agent in agent_docs:
            agent_docs[agent].add(doc)
        else:
            agent_docs[agent] = {doc}
    
   
    result = set()
    for agent, docs in agent_docs.items():
        if required_docs.issubset(docs):
            result.add(agent)
    
    return result

```

Задача 3. Обратный индекс
Обратный индекс - это структура, которая используется для поиска в массивах данных.

Вам нужно написать программу, которая принимает список документов (документ — это строка) и должна построить "обратный индекс". Это словарь, где:

ключ — это слово (в нижнем регистре).
значение — это множество индексов документов, в которых это слово встречается.
Пример:

docs = [
"Python is a great language",  # doc 0
"Java is also a great language", # doc 1
"Python and Java are popular"    # doc 2
]

Результат (порядок ключей не важен)

{
  "python": {0, 2},
  "is": {0, 1},
  "a": {0, 1},
  "great": {0, 1},
  "language": {0, 1},
  "java": {1, 2},
  "also": {1},
  "and": {2},
  "are": {2},
  "popular": {2}
}


```python
def index(docs):
    inverted_index = {}
    
    for doc_index, document in enumerate(docs):
        words = document.split()
        
        for word in words:
            word_lower = word.lower()
            
            if word_lower in inverted_index:
                inverted_index[word_lower].add(doc_index)
            else:
                inverted_index[word_lower] = {doc_index}
    
    return inverted_index
```
