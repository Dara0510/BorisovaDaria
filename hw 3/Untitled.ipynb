## Задание 1: Анализатор паролей (2 балла)

Напишите функцию `analyze_password`, которая проверяет надежность пароля.

**Что должна делать функция:**
1. Принимать строку с паролем
2. Проверять пароль по 5 критериям (каждый критерий = 1 балл силы):
   - Длина 8 символов или больше
   - Есть хотя бы одна заглавная буква (A-Z)
   - Есть хотя бы одна строчная буква (a-z)
   - Есть хотя бы одна цифра (0-9)
   - Есть хотя бы один специальный символ из набора: !@#$%^&*
3. Возвращать словарь с двумя ключами:
   - `'strength'` - число от 0 до 5 (сколько критериев выполнено)
   - `'suggestions'` - список строк с советами, что нужно добавить

**Пример работы:**
```python
analyze_password("abc")
# Вернет: {'strength': 1, 'suggestions': ['Увеличьте длину до 8 символов', 'Добавьте заглавные буквы', 'Добавьте цифры', 'Добавьте специальные символы']}
```

````python
def analyze_password(password):
    strength = 0
    suggestions = []
    
    if len(password) >= 8:
        strength += 1
    else:
        suggestions.append('Увеличьте длину до 8 символов')
    
    has_upper = False
    for char in password:
        if 'A' <= char <= 'Z':
            has_upper = True
            break
    if has_upper:
        strength += 1
    else:
        suggestions.append('Добавьте заглавные буквы')
    
    has_lower = False
    for char in password:
        if 'a' <= char <= 'z':
            has_lower = True
            break
    if has_lower:
        strength += 1
    else:
        suggestions.append('Добавьте строчные буквы')
    
    has_digit = False
    for char in password:
        if '0' <= char <= '9':
            has_digit = True
            break
    if has_digit:
        strength += 1
    else:
        suggestions.append('Добавьте цифры')
    
    special_chars = "!@#$%^&*"
    has_special = False
    for char in password:
        if char in special_chars:
            has_special = True
            break
    if has_special:
        strength += 1
    else:
        suggestions.append('Добавьте специальные символы')
    
    return {'strength': strength, 'suggestions': suggestions}
````

## Задание 4: Создатель персонажа игры с **kwargs (2 балла)

Создайте функцию `create_character`, которая создает персонажа для игры.

**Что должна делать функция:**
1. Принимать два обязательных параметра:
   - Имя персонажа (строка)
   - Класс персонажа (строка: 'warrior', 'mage' или 'archer')
2. Принимать любые дополнительные характеристики через именованные параметры
3. Автоматически добавлять базовые характеристики в зависимости от класса:
   - warrior: health=100, damage=15, armor=5
   - mage: health=60, damage=25, armor=2
   - archer: health=80, damage=20, armor=3
4. Если характеристика передана и как базовая, и через параметры - использовать переданную
5. Возвращать словарь со всеми характеристиками

**Примеры работы:**
```python
create_character('Артур', 'warrior', level=5)
# Вернет: {'name': 'Артур', 'class': 'warrior', 'health': 100, 'damage': 15, 'armor': 5, 'level': 5}

create_character('Мерлин', 'mage', health=70, mana=100)
# Вернет: {'name': 'Мерлин', 'class': 'mage', 'health': 70, 'damage': 25, 'armor': 2, 'mana': 100}
```

**Подсказка:** используйте **kwargs для приема именованных параметров

````python
def create_character(name, character_class, **kwargs):
    base_stats = {
        'warrior': {'health': 100, 'damage': 15, 'armor': 5},
        'mage': {'health': 60, 'damage': 25, 'armor': 2},
        'archer': {'health': 80, 'damage': 20, 'armor': 3}
    } 
    character = {
        'name': name,
        'class': character_class
    }
    for stat, value in base_stats[character_class].items():
        character[stat] = value
    character.update(kwargs)
    return character
````

1. Принимать список чисел
2. Принимать функцию-операцию (это может быть обычная функция или lambda)
3. Принимать необязательный параметр - функцию-фильтр (по умолчанию None)
4. Если фильтр задан:
   - Применять операцию только к элементам, для которых фильтр возвращает True
   - Остальные элементы оставлять без изменений
5. Если фильтр не задан - применять операцию ко всем элементам
6. Возвращать новый список


````python
def transform_list(numbers, operation, filter_func=None):
    result = []
    
    for num in numbers:
        if filter_func is None:
    
            new_num = operation(num)
            result.append(new_num)
        else: 
            if filter_func(num)      
                new_num = operation(num)
                result.append(new_num)
            else:
             
                result.append(num)
    
    return result
````
