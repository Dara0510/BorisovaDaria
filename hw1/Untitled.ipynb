Определите, сколько лет/месяцев/недель/дней/часов/минут в 5394523242 секундах.

```python
sec = 5394523242
min= sec// 60
hours = min// 60
days = hours // 24
week = days // 7
month = days // 30  
year = days // 365  

print(min)
print(hours)
print(days)
print(week)
print(month)
print(year)
```

Напишите код, который переводит целое число в строку, при том что его можно применить в любой системе счисления.

```python
def int_to_str(n, base=10):
    return str(n) if base == 10 else format(n, 'd')
```
