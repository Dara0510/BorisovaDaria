Вам нужно реализовать реализовывать алгоритм KNN (k nearest neighbours) «с нуля» для классификации, используя расстояние Евклида и Манхэттенское расстояние.

Как работает алгоритм:

- запоминает обучающую выборку
- при поступлении объекта для предсказания находит для этого объекта k ближайших соседей по метрике расстояния
- считает самый частотный класс из встретившихся среди объектов и выдает его как предсказание

**(5 баллов)** Вот шаблон класса, нужно обязательно реализовать все функции и (по желанию) поэкспериментировать с нормализацией и стандартизацией  признаков **(+ 1 балл**). Функции можно добавлять или модифицировать, но основная логика должна быть как в шаблоне ниже.

````python
import math
from collections import Counter

class KNearestNeighbors:
    
    def __init__(self, k=3):
        self.k = k
        self.x = None
        self.y = None
    
    def fit(self, x, y):
        self.x = x
        self.y = y
    
    def e_dist(self, a, b):
        s = 0
        for i in range(len(a)):
            s += (a[i] - b[i]) ** 2
        return math.sqrt(s)
    
    def m_dist(self, a, b):
        s = 0
        for i in range(len(a)):
            s += abs(a[i] - b[i])
        return s
    
    def pred(self, p, d='e'):
        all_d = []
        
        for i in range(len(self.x)):
            if d == 'e':
                dist = self.e_dist(p, self.x[i])
            else:
                dist = self.m_dist(p, self.x[i])
            all_d.append((dist, i))
        
        all_d.sort(key=lambda t: t[0])
        idx = [t[1] for t in all_d[:self.k]]
        lbl = [self.y[i] for i in idx]
        
        return Counter(lbl).most_common(1)[0][0]
````
