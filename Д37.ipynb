## 🚀 Подсказки и полезные примеры

### 1. Как переключаться между страницами?

Посмотрите на адреса страниц на Хабре:
- Первая страница: `https://habr.com/ru/all/`
- Вторая страница: `https://habr.com/ru/all/page2/`
- Пятидесятая страница: `https://habr.com/ru/all/page50/`

Мы можем легко создать нужный URL в цикле, просто подставляя номер страницы.

### 2. Как найти нужные данные на странице?

Откройте любую статью на Хабре, нажмите на заголовок правой кнопкой мыши и выберите "Исследовать элемент" (Inspect). Вы увидите HTML-код. Заголовки статей обычно находятся внутри тега `<a>` с классом `tm-title__link`. Имена авторов — в теге `<a>` с классом `tm-user-info__username`. Эти классы мы и будем использовать для поиска.

### 3. Как пользоваться библиотекой `tqdm`?

`tqdm` создает красивый индикатор выполнения. Достаточно "обернуть" в него любую последовательность, по которой вы итерируетесь (например, `range()`).

````python
print("Пример работы tqdm:")
for i in tqdm(range(10), desc="Имитация работы"):
    time.sleep(0.2) # Делаем вид, что что-то делаем
print("\nГотово!")
````

### 4. Как записывать данные в файл?

Для записи в файл лучше всего использовать конструкцию `with open(...)`. Она сама позаботится о закрытии файла. Не забывайте про `encoding='utf-8'` для корректной работы с русским языком.

````python
# Пример записи в файл
my_data = ["Строка 1", "Строка 2", "Строка 3"]

with open('example.txt', 'w', encoding='utf-8') as f:
    for line in my_data:
        f.write(line + '\n')

print("Данные записаны в example.txt")
````
